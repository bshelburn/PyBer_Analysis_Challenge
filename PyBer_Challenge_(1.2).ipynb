### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "/Users/benjaminshelburn/Desktop/Classwork/Week 5/PyBer_Analysis/city_data.csv"
ride_data_to_load = "/Users/benjaminshelburn/Desktop/Classwork/Week 5/PyBer_Analysis/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
pyber_data_df.dtypes
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type #showing wrong check variable
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type

total_driver_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]

total_driver_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
avgfares_perride_by_city_type = total_fares_by_city_type / total_rides_by_city_type
avgfares_perride_by_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
avgfare_perdriver_by_city_type = total_fares_by_city_type / total_drivers_by_city_type
avgfare_perdriver_by_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 

summary_df = pd.DataFrame()

summary_df["Total Rides"] = total_rides_by_city_type
summary_df["Total Drivers"] = total_drivers_by_city_type
summary_df["Total Fares"] = total_fares_by_city_type
summary_df["Average Fare per Ride"] = avgfares_perride_by_city_type
summary_df["Average Fare per Driver"] = avgfare_perdriver_by_city_type

summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index namesummary_df.index.name = None
total_fares_by_city_type.index.name = None
total_fares_by_city_type
```

```python
#  8. Format the columns.

summary_df["Total Rides"] = summary_df["Total Rides"].map("{:,}".format)
summary_df["Total Drivers"] = summary_df["Total Drivers"].map("{:,}".format)
summary_df["Total Fares"] = summary_df["Total Fares"].map("${:,.2f}".format)
summary_df["Average Fare per Ride"] = summary_df["Average Fare per Ride"].map("${:.2f}".format)
summary_df["Average Fare per Driver"] = summary_df["Average Fare per Driver"].map("${:.2f}".format)

summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

date_count = pyber_data_df.groupby(["date" ,"type"]).sum()["fare"]

date_count
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

date_count = date_count.reset_index()

date_count
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

pivot_table = pd.pivot_table(date_count, values='fare', index = 'date', columns ='type')

pivot_table
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

loc_pivottable = pivot_table.loc['2019-01-01':'2019-04-29']
loc_pivottable
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

loc_pivottable.index = pd.to_datetime(loc_pivottable.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()

loc_pivottable.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

loc_pivottable_res=loc_pivottable.resample('W').sum()

loc_pivottable_res.info()
```

```python
loc_pivottable_res
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

loc_pivottable_res.plot(figsize=(20,10))
plt.ylabel("Fares($)")
plt.xlabel("Month")
plt.title("Total Fares by City Type")

```
